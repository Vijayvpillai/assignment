```python
# no:1

n = int(input("enter a number"))
j = 1
for i in range(1,2*n) :
    if i <= n :
        print("* " * i)
    else :
        i = i-2*j
        print("* " * i)
        j = j+1
    
```

```python
# no:2.1

name = input("enter something \n")
for i in range(len(name)) :
    j = i+1
    print(name[(len(name))-j],end='')
```
